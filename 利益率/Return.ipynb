```python
#pandas_datareaderのインポート
import pandas_datareader.data as wb
```

```python
#データ読み込み
df=wb.DataReader('9984.T','yahoo',start='2020-1-1',end='2021-12-31')
df.head()
```

```python
#単純利益率の計算
df['simple_return']=(df['Adj Close']/df['Adj Close'].shift(1))-1
df['simple_return'].head(10)
```

```python
#図の描画
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(10,4))
plt.plot(df.index,df['simple_return'])
plt.savefig('./image/fig_simple_return.jpg')
plt.show()
```

```python
#1日あたりの利益率
avg_return_d=df['simple_return'].mean()
#1年あたりの利益率
avg_return_a=avg_return_d*250

print(str(round(avg_return_a,5)*100)+'%')
```

```python
#対数利益率の計算
import numpy as np

df['log_return']=np.log(df['Adj Close']/df['Adj Close'].shift(1))
df['log_return'].head(10)
```

```python
#図の描画
plt.figure(figsize=(10,4))
plt.plot(df.index,df['log_return'])
plt.savefig('./image/fig_log_return.jpg')
plt.show()
```

```python
#1日あたりの利益率
avg_return_d=df['log_return'].mean()
#1年あたりの利益率
avg_return_a=avg_return_d*250

print(str(round(avg_return_a,5)*100)+'%')
```
